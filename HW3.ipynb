# **LINK 1 CLEANING & FORMATTING**

```r
#creates a variable that stores a web link for that variable
HW3LINK1="https://github.com/DACSS601HW/HW3/raw/refs/heads/main/HW3.1_Maternal%20mortality%20ratio.csv"
```

```r
#This line downloads a CSV file from the web and loads it into R as a data frame, storing it in a variable called DIRTY1
DIRTY1=read.csv(HW3LINK1)
```

```r
#inspects the internal structure of the dataset so you can quickly understand what kind of data you’re working with
str(DIRTY1)
```

```r
#displays the first few rows of DIRTY1
head(DIRTY1)
```

```r
#Displays the last few rows of DIRTY1
tail(DIRTY1)
```

```r
#Displays column names of DIRTY1
names(DIRTY1)
```

```r
#removes unwanted columns from the data frame
DIRTY1 <- DIRTY1[, !(names(DIRTY1) %in% c('region', 'X000.live.births', 'date_of_information', 'slug'))]
```

```r
#checks the internal structure of the object named DIRTY1
str(DIRTY1)
```

```r
#This code finds three specific columns in DIRTY1 and replaces their names with more readable labels
names(DIRTY1)[names(DIRTY1) %in% c("name", "deaths.100", "ranking")] <-
  c("Country", "Deaths per 100k Live Births", "Region")
```

```r
#converts all values in the Country column to uppercase text
DIRTY1$Country <- toupper(DIRTY1$Country)
```

```r
#Displays all items in the dataframe (Object) DIRTY1
DIRTY1
```

```r
#saves your cleaned R data frame to disk in R’s native .RDS format
saveRDS(DIRTY1,"HW3.1_MaternalMortalityRatio.RDS")
```

```r
#loading back the saved .RDS file and inspecting its structure
DIRTY1RDS=readRDS("HW3.1_MaternalMortalityRatio.RDS")
str(DIRTY1RDS)
```

```r
#saving cleaned data to a CSV file, reloading it, and checking its structure
write.csv(DIRTY1,"HW3.1_MaternalMortalityRatio.csv", row.names=FALSE)
DIRTY1CSV=read.csv("HW3.1_MaternalMortalityRatio.csv")
str(DIRTY1CSV)
```
